* Se basa en la ecuacion y=ax + b

```python
#Importando librerias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets,linear_model
from sklearn.model_selection import train_test_split
```

```python
#IMPORTANDO LOS DATOS
peso_estat = pd.read_csv("C:/Users/Arley/Desktop/AMachine Learning/Modelos MachineLearning/nutricion.csv", sep=";")

#SELECCIONANDO LOS DATOS A USAR
datos = peso_estat[["peso","estatura"]][0:1000]

#Analizando si faltan datos
datos.notnull().count()

```

```python
#GRAFICANDO DATOS
x_data = datos["peso"]
y_data = datos["estatura"]

plt.scatter(x_data, y_data, color='blue', linewidth=3)
plt.title("Peso vs estatura en niños")
plt.ylabel("Altura cm")
plt.xlabel("Peso kg")
plt.show
```

## Observaciones 
* En el grafico se observa una correlacion claramente lineal

# MODELO REGRESION

```python
# Separamos los datos de entrenamiento de los de prueba
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.5)

#Modificando los datos ya que el modelo lo pide en mas de una dimension [[1D],[2D]]
x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)

x_train = x_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)

```

```python
#Se define el modelo a usar
lr = linear_model.LinearRegression()
```

```python
#ENTRENANDO EL MODELO
lr.fit(x_train, y_train)
```

```python
#HACIENDO LAS PREDICCIONES
y_predic = lr.predict(x_test)
```

```python
#Analizando los datos
#GRAFICANDO DATOS
plt.scatter(x_train, y_train, color='blue', linewidth=2, label="Datos de Entrenamiento")
plt.plot(x_test, y_predic, color='red', linewidth=2, label="Datos de Prediccion")
plt.title("Peso vs estatura en niños")
plt.legend()
plt.ylabel("Altura cm")
plt.xlabel("Peso kg")
plt.show
```

```python
#Obteniendo la ecuacion
lr.coef_ #Valor de la pendiente
lr.intercept_ #Valor del intercepto

#ECUACION 
print(f"y = {lr.coef_}X + {lr.intercept_}")
```

# PRECISION DEL MODELO

```python
lr.score(x_train,y_train)
```
